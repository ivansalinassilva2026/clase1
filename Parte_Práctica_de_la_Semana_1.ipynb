### 4. Indexación

En diversos tipos de datos de Python es posible acceder directamente a elementos individuales dentro de una estructura ordenada mediante un índice numérico o, en otros casos, a través de una clave. Este mecanismo se conoce como indexación.

En el caso de las cadenas de texto, Python las trata como secuencias ordenadas de caracteres, por lo que cada uno de sus elementos puede ser recuperado individualmente mediante su posición. Para ello, se escribe el nombre de la variable seguido del índice correspondiente entre corchetes [].

La indexación en Python comienza en 0, lo que significa que el primer carácter de una cadena se encuentra en la posición 0, el segundo en la posición 1, y así sucesivamente. En consecuencia, el índice del último carácter corresponde a la longitud total de la cadena menos uno.

````python
nombre = "Francisco"
````

````python
nombre[0]
````

````python
nombre[4]
````

````python
nombre[-1]
````

````python
nombre[-3]
````

### 5. Slicing

Python también incorpora una notación específica de indexación denominada slicing, la cual permite extraer subsecuencias o subcadenas a partir de una cadena de texto. Este mecanismo resulta especialmente útil para manipular segmentos de información textual de manera eficiente y controlada.

La sintaxis general del slicing sobre una cadena s se expresa como s[m:n], donde m representa el índice inicial y n el índice final del segmento que se desea obtener. El resultado corresponde a la porción de la cadena que comienza en la posición m y se extiende hasta la posición n-1, ya que el índice superior no se incluye en la extracción.

````python
nombre = "Francisco Escudero"
````

````python
nombre[0:9]
````

````python
nombre[-12:1]
````

Si se omite uno de los índices en la operación de slicing, Python asume automáticamente el límite correspondiente de la secuencia; es decir, desde el inicio si no se especifica el primer valor, o hasta el final si no se indica el segundo.

````python
nombre[-9:]
````

````python
nombre[4:]
````

### 6. Stride

El stride es una variante avanzada del slicing que permite controlar el intervalo entre los elementos extraídos de una secuencia. En la expresión s[inicio:fin:paso], el tercer parámetro, denominado stride o paso, indica cada cuántos caracteres se seleccionará un nuevo elemento dentro de la cadena. De este modo, es posible recorrer la secuencia de forma no contigua, extrayendo caracteres según la separación especificada.

````python
nombre = "Francisco Escudero"
````

````python
nombre[0:8:2]
````

# **III. Funciones**

En el ámbito matemático, una función se define como una relación que asocia una o más entradas con una salida determinada, y suele representarse de forma general como y = f(x). No obstante, en programación, el concepto de función adquiere un alcance más amplio y operativo. En este contexto, una función corresponde a una unidad lógica de código que encapsula una tarea específica o un conjunto de operaciones relacionadas, permitiendo su reutilización, organización y abstracción dentro de un programa. Su uso favorece el diseño modular, ya que facilita la descomposición de problemas complejos en componentes más pequeños, comprensibles y mantenibles, aspecto esencial en el desarrollo de algoritmos, sistemas de análisis de datos y aplicaciones de inteligencia artificial.

````python
var = "Ciencia de Datos"
````

````python
len(var)
````

La función presentada anteriormente forma parte del conjunto de funciones integradas que ofrece el intérprete de Python, las cuales pueden ser utilizadas directamente siempre que resulte necesario. Para hacer uso de estas funciones no es indispensable conocer el detalle de su implementación interna, sino comprender adecuadamente su interfaz, es decir, qué argumentos reciben y qué tipo de valor retornan.

Cuando una función es invocada mediante su nombre y se le proporcionan los argumentos requeridos, el intérprete de Python ejecuta internamente las instrucciones asociadas a su definición, aplicando las operaciones correspondientes sobre los datos de entrada. Una vez finalizada su ejecución, el control del programa retorna al punto exacto desde el cual la función fue llamada. Dependiendo de su diseño, la función puede además entregar un valor de retorno, el cual puede ser almacenado, procesado o reutilizado en etapas posteriores del código.

### 1. Definiendo funciones personalizadas

La sintaxis utilizada para definir funciones en Python es la siguiente:


````python
def <nombre_funcion>([<parámetros>]):
    <sentencia(s)>
````

El último elemento, `<sentencia(s)>`, se denomina cuerpo de la función. El cuerpo es el bloque de sentencias en Python que se ejecutará cuando se llame a la función. El cuerpo de una función de Python se define por la sangría.

````python
def mi_funcion(arg1, arg2):
    print(arg1)
    print(arg2)
````

En Python, la llamada a una función se realiza mediante la sintaxis general `<nombre_funcion>([<argumentos>])`. En esta expresión, los argumentos representan los datos concretos que se entregan en el momento de la invocación, mientras que los parámetros corresponden a las variables formales definidas en la estructura de la función para recibir dichos valores. Esta correspondencia entre argumentos y parámetros permite ejecutar procedimientos de manera flexible, modular y reutilizable dentro del programa.

````python
mi_funcion("Hola mundo", "Adios mundo")
````

````python
print(1)
mi_funcion(2, 3)
print(4)
````

Python permite la definición de funciones sin parámetros y su correspondiente invocación sin argumentos; no obstante, en ambos casos el uso de paréntesis es obligatorio. Esto se debe a que los paréntesis constituyen un elemento sintáctico fundamental en la declaración y llamada de funciones, aun cuando no exista información que transferir entre la función y el contexto desde el cual se ejecuta.

````python
def mi_funcion2():
    print("Ciencia de Datos")
````

````python
mi_funcion2()
````

### 2. Argumentos de las funciones

En Python, los argumentos que se proporcionan al invocar una función no están restringidos a un único tipo de dato, sino que pueden pertenecer a diversas categorías, incluyendo tipos primitivos y estructuras más complejas. Esta característica aporta versatilidad al diseño funcional del lenguaje y facilita la construcción de soluciones modulares aplicables a distintos contextos de programación, análisis de datos e inteligencia artificial.

### 2.1 Argumentos posicionales

En Python, una de las formas más básicas de transferir información a una función es a través de argumentos posicionales. En este mecanismo, la correspondencia entre argumentos y parámetros se establece según la posición que ocupan en la llamada y en la definición de la función, respectivamente. Por esta razón, la función debe declarar explícitamente, entre paréntesis y separados por comas, los parámetros que definirán la estructura esperada de entrada.

````python
def mi_funcion(arg1, arg2, arg3):
    print(arg1)
    print(arg2)
    print(arg3)
````

El número de argumentos que le proporcionamos en la invocación de la función debe respetar el orden y el número de parámetros definidos.

````python
mi_funcion("Magister", "Ciencia de Datos", "Inteligencia Artificial")
````

````python
mi_funcion("Magister", "Ciencia de Datos")
````

En Python, los parámetros constituyen variables locales asociadas exclusivamente al entorno interno de la función, por lo que su alcance queda restringido al bloque de código que conforma su definición. Durante la invocación, los argumentos suministrados son enlazados a los parámetros conforme al orden en que estos fueron declarados, estableciendo la correspondencia necesaria para la ejecución del procedimiento definido.

````python
def mi_funcion():
    var = "variable 'var' dentro de la funcion"
    print(var)
````

````python
var = "variable 'var' fuera de la funcion"
mi_funcion()
print(var)
````

### 2.2 Argumentos de palabra clave


En Python, una función puede invocarse mediante argumentos nombrados, utilizando la estructura `<palabra_clave>=<valor>`. Bajo este mecanismo, cada palabra clave debe coincidir con un parámetro formal declarado en la definición de la función, estableciendo una correspondencia explícita entre los valores suministrados y las variables receptoras. Esta modalidad mejora la legibilidad del código y aporta mayor flexibilidad en la llamada a funciones con múltiples parámetros.



````python
def mi_funcion(arg1, arg2):
    print(arg1)
    print(arg2)
````

````python
var = "Ciencia de Datos"
var2 = "Inteligencia Artificial"

mi_funcion(arg1=var, arg2=var2)
````

````python
mi_funcion(arg1=var, arg3=var2)
````

La utilización de argumentos nombrados en Python elimina la dependencia del orden posicional en la llamada a una función, puesto que la asignación entre argumentos y parámetros se establece de manera explícita a través del nombre de cada parámetro.

````python
mi_funcion(arg1=var, arg2=var2)
````

````python
mi_funcion(arg2=var2, arg1=var)
````

Si bien el uso de argumentos nombrados otorga flexibilidad respecto del orden de asignación, no elimina la necesidad de satisfacer la aridad de la función; es decir, debe proporcionarse el número de argumentos exigido por su definición, excepto en aquellos casos en que existan parámetros opcionales o estructuras variables de entrada.

````python
mi_funcion(var, arg2=var2)
````

### 2.3 Parámetros con valores por defecto u opcionales

En Python, cuando un parámetro se define utilizando la sintaxis `<nombre>=<valor>`, el `<valor>` asignado queda establecido como valor por defecto dentro de la firma de la función. Esto implica que su provisión en la llamada deja de ser obligatoria, ya que, en ausencia de un argumento explícito, el intérprete empleará automáticamente el valor predeterminado. Este tipo de parámetros se denomina parámetros opcionales o parámetros con valor por defecto.

````python
def mi_funcion(arg1, arg2="Inteligencia Artificial"):
    print(arg1)
    print(arg2)
````

````python
mi_funcion("Ciencia de Datos")
````

````python
mi_funcion("Inteligencia Artificial", "Ciberseguridad")
````

### 3. Sentencia Return

En Python, una función puede devolver un valor una vez finalizada la ejecución de las instrucciones que conforman su cuerpo. Para ello, se utiliza la palabra clave `return`, cuya finalidad es transferir un resultado desde el ámbito interno de la función hacia el contexto desde el cual esta fue invocada. De este modo, el valor retornado puede ser almacenado, reutilizado en otras expresiones o incorporado en etapas posteriores del flujo computacional.

Al utilizar la instrucción `return`, es importante considerar varios aspectos relacionados con el comportamiento de la función y con la forma en que dicho valor de salida será gestionado dentro del programa.

3.1 La ejecución de la sentencia `return` provoca la terminación inmediata del flujo interno de la función, transfiriendo el control nuevamente a la línea de código donde se realizó la llamada, junto con el valor de retorno correspondiente, si este ha sido especificado.

````python
def mi_funcion():
    print("Sentencia 1 en mi_funcion")
    return
    print("Sentencia 2 en mi_funcion")
````

````python
print("Ciencia de Datos")
mi_funcion()
print("Ciberseguridad")
````

3.2 Cuando la sentencia `return` adopta la estructura `return` `<valor>`, el intérprete finaliza la ejecución de la función y transfiere el `<valor>` especificado como resultado de salida hacia la llamada que originó su ejecución.

````python
def mi_funcion():
    return "Valor de mi_funcion"
````

````python
var = mi_funcion()

print(var)
````

````python
mi_funcion()
````

````python
mi_funcion()[0:5]
````

3.3 Cuando la instrucción `return` no incluye un `<valor>` asociado, Python retorna por defecto `None`, un objeto especial que indica que la función ha finalizado sin producir un resultado explícito.

````python
def mi_funcion():
    return
````

````python
var = mi_funcion()
````

````python
print(var)
````

3.4 Python permite que una función retorne múltiples valores mediante una única sentencia `return`, escribiendo los elementos separados por comas. Aunque sintácticamente se perciben como valores independientes, el intérprete los encapsula internamente en una tupla, lo que facilita su asignación simultánea a múltiples variables o su tratamiento como una estructura compuesta de salida. Devolviendo varios valores utilizando la sintaxis `return <valor1>,<valor2>,...,<valorn>`

````python
def mi_funcion():
    return "texto1","texto2","texto3"
````

````python
var1,var2,var3 = mi_funcion()
````

````python
print(var1)
print(var2)
print(var3)
````

### 4. Docstrings

Cuando la primera instrucción dentro del cuerpo de una función en Python corresponde a una cadena de texto delimitada, por ejemplo, mediante comillas triples """texto""", esta se interpreta como el docstring de la función.

El docstring cumple una función documental, ya que permite describir formalmente el propósito de la función, los parámetros que recibe, los valores que retorna y cualquier otra información relevante para su correcta comprensión y uso. Su incorporación favorece la legibilidad, mantenibilidad y reutilización del código, especialmente en proyectos de mayor complejidad o desarrollados de forma colaborativa.

Existe un PEP (Python Enhancement Proporsal) que establece el estilo con el que deben definirse los docstrings: https://www.python.org/dev/peps/pep-0257/

````python
def mi_funcion():
    """Esto es un docstring de la función."""
    print("Ciencia de Datos")
````

````python
mi_funcion()
````

````python
def mi_funcion2(arg1=0.0, arg2=0.0):
    """Esto es un docstring de varias líneas.

    Keyword arguments:
    arg1 -- primer argumento de la funcion (default 0.0)
    arg2 -- segundo argumento de la funcion (default 0.0)
    """
    return
````

````python
mi_funcion2()
````

# 2. Tuplas

En Python, una tupla es una estructura secuencial ordenada que comparte varias propiedades con las listas, pero difiere de ellas principalmente en su sintaxis y en su comportamiento frente a la modificación. Las tuplas se definen mediante paréntesis, siguiendo la forma `(obj1, obj2, ..., objn)`, y se caracterizan por ser inmutables, es decir, no admiten alteraciones en sus elementos después de su instanciación. Esta propiedad las hace especialmente útiles en contextos donde se requiere preservar la integridad de los datos.

````python
tupla = (1, 2, 3, 4, 5)
````

````python
tupla[2] = "text0"
````

¿Cuándo utilizar una tupla en lugar de una lista?
Desde una perspectiva de diseño y eficiencia computacional, las tuplas resultan preferibles a las listas en contextos donde la colección de datos no requiere modificaciones posteriores a su creación. En primer lugar, las tuplas suelen presentar un rendimiento ligeramente superior en términos de almacenamiento y acceso, debido a su naturaleza inmutable y a una implementación más liviana que la de las listas, aunque esta diferencia adquiere mayor relevancia en estructuras de datos de tamaño considerable. En segundo lugar, la inmutabilidad de las tuplas proporciona una garantía adicional de integridad, ya que evita alteraciones accidentales sobre los valores almacenados, aspecto especialmente importante en el desarrollo de programas robustos y mantenibles. Finalmente, ciertos componentes del lenguaje, como las claves de los diccionarios, exigen objetos inmutables y hashables; en este escenario, las tuplas pueden utilizarse válidamente, siempre que sus elementos también sean inmutables, mientras que las listas quedan excluidas debido a su carácter mutable.

Se debe tener especial cuidado al definir una tupla de un solo elemento cuyo contenido sea numérico, ya que Python puede interpretar los paréntesis únicamente como un mecanismo de agrupación de una expresión aritmética, y no como la construcción de una tupla. En consecuencia, para que el intérprete reconozca correctamente una tupla unitaria, es necesario incluir una coma después del único elemento, por ejemplo: (5,). De lo contrario, una expresión como (5) será tratada simplemente como un valor numérico y no como una tupla.

````python
tupla = (2)
````

````python
tupla
````

````python
type(tupla)
````

La forma correcta de resolver esta ambigüedad sintáctica consiste en definir la tupla mediante la expresión (num,), ya que la presencia de la coma es el elemento que permite a Python reconocer la estructura como una tupla unitaria. Sin dicha coma, el valor sería interpretado únicamente como un objeto numérico agrupado entre paréntesis.

````python
tupla = (2,)
````

````python
tupla
````

````python
type(tupla)
````

Este mecanismo, denominado unpacking, facilita una manipulación más clara, concisa y expresiva de los datos.

En consecuencia, al escribir una sentencia como return valor1, valor2, valor3, el intérprete construye implícitamente una tupla con esos elementos y la utiliza como valor de retorno de la función. # 3. Sentencia `while`

La sentencia while constituye otra de las estructuras fundamentales de control de flujo en Python. Su propósito es ejecutar de manera repetida un bloque de código mientras una condición lógica determinada permanezca verdadera. A diferencia de la sentencia for, que suele emplearse cuando el número de iteraciones es conocido o finito, while resulta especialmente útil en situaciones donde la repetición depende de una condición dinámica evaluada durante la ejecución del programa.

En este sentido, la sentencia while puede dar lugar a un número indeterminado de iteraciones y, si la condición nunca deja de cumplirse, incluso a un bucle infinito. Su sintaxis general en Python es la siguiente:


```
while <expresión>:
    <sentencia(s)>
```
  
`<sentencia(s)>` es el bloque de sentencias en Python que se ejecutará de manera indefinida hasta que no se cumpla la expresión.

`<expresión>` es una expresión evaluada en un contexto booleano. El resultado de evaluar esta expresión debe ser `True` o `False`. Controla la ejecución del bucle `while`.

````python
num = 10
while num > 0:
    num -= 1
    print(num)
````

La sentencia break es una palabra reservada de Python que permite interrumpir de forma inmediata la ejecución de un bucle for o while. Cuando esta instrucción es alcanzada, el flujo de control abandona la estructura iterativa en curso y se transfiere directamente a la primera línea de código ubicada fuera del bloque correspondiente. Este mecanismo resulta especialmente útil para finalizar anticipadamente una iteración cuando se satisface una condición específica o cuando continuar con el bucle deja de ser necesario.

````python
colores = ["rojo", "verde", "azul"]
for color in colores:
    print(color)
````

````python
for color in colores:
    print(color)
    break
````

````python
for color in colores:
    print(color)
    break
else:
    print("Ejecución en la claúsula else")
````

# 5. Sentencia `continue`

La sentencia continue es otra palabra reservada de Python que, a diferencia de break, no interrumpe por completo la ejecución de un bucle for o while. En lugar de ello, finaliza únicamente la iteración actual y transfiere el control al siguiente ciclo de evaluación del bucle. Este comportamiento resulta especialmente útil cuando se desea omitir determinadas condiciones o casos específicos sin detener por completo la estructura iterativa.

````python
colores = ["azul", "verde", "rojo"]
for color in colores:
    print(color)
````

````python
for color in colores:
    if color == "verde":
        continue
    print(color)
````

# **VI. Programación Modular**


La programación modular constituye un principio fundamental en el diseño de software contemporáneo, especialmente en contextos donde la complejidad computacional, la mantenibilidad del código y la escalabilidad de las soluciones adquieren un rol central. Este enfoque propone descomponer un sistema en unidades funcionales más pequeñas, cohesionadas y relativamente independientes, denominadas módulos, cada una de las cuales asume una responsabilidad específica dentro de la solución global. En contraposición a una arquitectura monolítica, en la que la lógica del programa se concentra en un único bloque de código, la modularización permite estructurar el desarrollo en componentes que pueden ser diseñados, implementados, evaluados y reutilizados de manera separada, para luego integrarse en una arquitectura común y consistente. Desde esta perspectiva, la modularidad no solo favorece la claridad conceptual del sistema, sino que también mejora su capacidad de adaptación, evolución y trazabilidad a lo largo del ciclo de vida del software.

Desde un punto de vista técnico, la modularización ofrece múltiples beneficios. En primer lugar, contribuye a la simplicidad estructural, al permitir abordar problemas complejos mediante una estrategia de descomposición en subproblemas más acotados y manejables. En conjunto, estas construcciones favorecen el desarrollo de sistemas más robustos, mantenibles y escalables, una condición particularmente relevante en proyectos de ciencia de datos e inteligencia artificial, donde los flujos de procesamiento suelen involucrar ingestión de datos, transformación, modelamiento, evaluación, visualización y despliegue, todos ellos susceptibles de beneficiarse de una arquitectura modular.

# VII. Excepciones

En Python, los errores que aparecen durante la ejecución de un programa se gestionan mediante un mecanismo llamado excepciones. Una excepción es un objeto que representa una situación anómala o inesperada, como intentar dividir por cero, acceder a un elemento inexistente o abrir un archivo que no se encuentra disponible. Cuando ocurre uno de estos eventos, Python genera automáticamente una excepción e interrumpe el flujo normal de ejecución.

Si el programa incluye instrucciones para manejar esa situación, la excepción puede capturarse y tratarse de forma controlada, permitiendo que la ejecución continúe o que el sistema responda adecuadamente al problema. En cambio, si no existe un manejo explícito, el programa se detiene y Python muestra un mensaje con información sobre el error ocurrido. Desde una perspectiva de desarrollo de software, este mecanismo es esencial para construir aplicaciones más robustas, seguras y tolerantes a fallos.

En Python, las excepciones se gestionan principalmente mediante las sentencias try y except. El bloque try contiene el código que podría generar un error, mientras que el bloque except define cómo actuar si dicho error ocurre. De esta forma, el manejo de excepciones no solo permite detectar fallos, sino también controlar de manera más precisa el comportamiento del programa ante situaciones imprevistas.

### 1. Errores de sintaxis vs Excepciones

A pesar de que en muchos contextos estos términos suelen emplearse de manera indistinta, en Python es importante distinguir entre errores de sintaxis y excepciones, ya que corresponden a situaciones diferentes dentro del proceso de ejecución de un programa.

Los errores de sintaxis se producen cuando el código escrito no respeta las reglas formales del lenguaje. En estos casos, el problema no está en la lógica de ejecución, sino en que la instrucción simplemente no puede ser interpretada correctamente por Python. Cuando esto ocurre, el intérprete genera un mensaje identificado como SyntaxError y suele señalar la ubicación aproximada del problema mediante el carácter ^, facilitando así la detección del fragmento de código inválido.

A diferencia de las excepciones, los errores de sintaxis se detectan antes de que el programa pueda ejecutarse normalmente. Por ello, constituyen fallos en la escritura del código que deben corregirse para que la ejecución sea posible.

````python
print("Hola mundo)
````

Los errores de sintaxis no pueden ser gestionados mediante mecanismos de control de excepciones, ya que no corresponden a excepciones propiamente tales dentro del lenguaje. Estos errores se producen cuando el código escrito no respeta las reglas formales de la sintaxis de Python, por lo que el intérprete no puede siquiera comenzar a ejecutarlo correctamente. En cambio, las excepciones se originan durante la ejecución de sentencias que sí son sintácticamente válidas, pero que generan una situación anómala, como una división por cero o el intento de acceder a un recurso inexistente. A diferencia de los errores de sintaxis, las excepciones no necesariamente implican la detención definitiva del programa, ya que pueden ser capturadas, gestionadas o incluso ignoradas mediante las estructuras de control apropiadas.

### 2. Gestión de excepciones: Sentencias `try` y `except`

Las sentencias `try` y `except` en Python pueden utilizarse para capturar y manejar excepciones. Python ejecuta el código que sigue a la sentencia `try` como una parte "normal" del programa. El código que sigue a la sentencia `except` se ejecutará si se produce cualquier excepción en la cláusula `try` precedente.

La sintaxis que se utiliza para definir este tipo de comportamiento es la siguiente:
```
try:
    <sentencia(s)>
except <excepción>:
    <sentencias(s) si excepción>
```

````python
print(var)
````

````python
try:
    print(var)
except NameError:
    var = "Hola mundo"
````

````python
print(var)
````

Debemos tener en cuenta que, si dentro del bloque try se produce una excepción, las sentencias ubicadas después de la instrucción que la genera no llegarán a ejecutarse. En ese caso, el flujo del programa se transfiere inmediatamente al bloque except correspondiente.

````python
print("Sentencias de codigo antes del try")

try:
    print("Codigo antes de la excepcion")
    10 + "3"
    print("Codigo despues de la excepcion")
except TypeError:
    print("No se puede sumar un numero entero y un string")

print("Sentencias de codigo despues del except")
````

También podemos utilizar la sentencia `except` sin indicarle el nombre de ninguna excepción, en estos casos capturará todas las excepciones que se produzcan en el código que se encuentra en el cuerpo de la sentencia `try`

````python
try:
    200 + "10"
except:
    print("No sumar un entero con un string")
````

Además de la sintaxis básica, es posible capturar distintos tipos de excepciones mediante el uso de múltiples cláusulas except. Esto permite definir un tratamiento específico para cada tipo de error, lo que proporciona un mayor control sobre el comportamiento del programa ante distintas situaciones excepcionales.

````python
try:
    50/0
except NameError:
    print("Gestionando excepcion NameError")
except TypeError:
    print("Gestionando execpcion Type Error")
except ZeroDivisionError:
    print("No puedes dividir un numero por 0")
````

Finalmente, Python permite asociar el objeto de excepción capturado a una variable, lo que hace posible acceder a su contenido y utilizarlo para proporcionar información más específica sobre el error producido.

````python
try:
    print(variable)
except NameError as error:
    print("Objeto de tipo:", type(error))
    print("La excepcion consiste en:", error)
````

### 1.1 Array

Un array es una estructura de datos que permite almacenar múltiples elementos del mismo tipo de forma organizada. En NumPy, los arrays pueden tener una o varias dimensiones, lo que los convierte en una herramienta central para representar vectores, matrices y estructuras de datos más complejas. Cada posición dentro del array se identifica mediante uno o más índices, lo que permite acceder de manera precisa a sus elementos. Esta forma de organización resulta especialmente útil en ciencia de datos, ya que facilita operaciones numéricas eficientes sobre grandes volúmenes de información.

En el contexto de NumPy, cada dimensión del array se denomina eje (axis). El número total de dimensiones se conoce como número de dimensiones o rank. La forma (shape) corresponde a una tupla que indica cuántos elementos hay en cada dimensión, mientras que el tamaño (size) representa el número total de elementos almacenados en el array. Estas propiedades permiten describir con precisión la estructura del arreglo y comprender cómo se distribuyen sus datos.

````python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])

print(a)
print("Número de dimensiones:", a.ndim)
print("Forma:", a.shape)
print("Tamaño:", a.size)
````

````python
# Array cuyos valores son todos 0
a = np.zeros((2, 4))
````

En este ejemplo, el array tiene dos ejes: uno correspondiente a las filas y otro a las columnas. Su rank es 2, porque posee dos dimensiones. Su shape es (2, 3), lo que indica que tiene 2 filas y 3 columnas. Finalmente, su size es 6, porque contiene 6 elementos en total.

Una forma breve de resumirlo sería esta:

* Axis: cada dimensión del array

* Rank / ndim: cantidad de dimensiones

* Shape: tamaño de cada dimensión

* Size: número total de elementos

### 3. Codificación de variables categóricas

En ciencia de datos, muchas variables representan **categorías** y no cantidades numéricas. Estas se denominan **variables categóricas** y describen cualidades, clases o grupos, como por ejemplo `color`, `tipo de cliente` o `nivel de satisfacción`.

Dado que la mayoría de los algoritmos de aprendizaje automático requieren entradas numéricas, es necesario aplicar un proceso de **codificación** (*encoding*), es decir, transformar estas categorías en una representación numérica adecuada para el análisis y el modelamiento.

### ¿Qué es una variable categórica?

Una **variable categórica** es aquella cuyos valores representan clases o categorías, y no cantidades con significado aritmético. Dentro de este tipo de variables, es posible distinguir dos casos principales:

### a. Variables categóricas nominales

Las **variables nominales** son aquellas cuyas categorías **no presentan un orden natural** entre sí.

**Ejemplo:**
- color = {rojo, azul, verde}

En este caso, las categorías son diferentes, pero ninguna es superior o inferior a otra. Por ello, no es adecuado asignar códigos como:

- rojo = 1  
- azul = 2  
- verde = 3  

Esto podría llevar al modelo a interpretar erróneamente que existe una relación de orden entre las categorías.

#### ¿Qué técnica se utiliza?
Para este tipo de variables se emplea **One-Hot Encoding**, que transforma cada categoría en una columna binaria independiente.

**Ejemplo:**

- rojo → [1, 0, 0]  
- azul → [0, 1, 0]  
- verde → [0, 0, 1]  

De esta manera, se representa la pertenencia a una categoría sin imponer una jerarquía artificial.

### b. Variables categóricas ordinales

Las **variables ordinales** son aquellas cuyas categorías **sí poseen un orden o jerarquía natural**.

**Ejemplo:**
- nivel de satisfacción = {bajo, medio, alto}

Aquí sí existe una progresión lógica entre las categorías, por lo que es razonable asignar una codificación numérica que preserve ese orden.

#### ¿Qué técnica se utiliza?
Para este tipo de variables se utiliza **Label Encoding** o codificación ordinal.

**Ejemplo:**

- bajo = 0  
- medio = 1  
- alto = 2  

En este caso, los números no representan magnitudes exactas, pero sí mantienen la relación jerárquica entre las categorías.

### ¿Por qué se usa una técnica distinta en cada caso?

La elección del método de codificación depende de la naturaleza de la variable:

- Para variables **nominales**, se utiliza **One-Hot Encoding** porque evita crear relaciones de orden inexistentes.
- Para variables **ordinales**, se utiliza **Label Encoding** porque conserva la jerarquía natural entre las categorías.

Esta distinción es metodológicamente importante, ya que una codificación incorrecta puede afectar la forma en que el modelo interpreta los datos y, en consecuencia, deteriorar la calidad de los resultados.

### Resumen

- **Variable categórica**: representa clases o categorías.
- **Variable nominal**: no tiene orden natural.
- **Variable ordinal**: tiene una jerarquía entre sus categorías.
- **One-Hot Encoding**: se utiliza para variables nominales.
- **Label Encoding**: se utiliza para variables ordinales.

> **Nota:** En `scikit-learn`, `LabelEncoder` se utiliza principalmente para codificar la variable objetivo. Para variables ordinales de entrada suele preferirse `OrdinalEncoder`. Sin embargo, conceptualmente ambos responden a la misma idea: representar categorías mediante códigos numéricos respetando su orden.

PARTE 1 — IMPLEMENTACIÓN MANUAL
Obtención y Exploración de Datos
Exploración manual del dataset para comprender su estructura.

````python
# -----------------------------------------
# FUNCIÓN: cargar_csv
# -----------------------------------------
# Esta función representa la etapa de INGESTA DE DATOS.
#
# Objetivo:
# - Leer manualmente un archivo CSV.
# - Convertir cada fila en un diccionario.
# - Manejar valores faltantes como None.
#
# Importante:
# No usamos pandas. Todo se hace manualmente
# para comprender cómo funciona internamente
# la lectura estructurada de datos.


def cargar_csv(ruta):

    """
    Lee manualmente un archivo CSV.
    Maneja valores vacíos como None.

    Parámetro:
    ruta (str): Ruta del archivo CSV.

    Retorna:
    encabezados (list): Lista de nombres de columnas.
    datos (list): Lista de diccionarios (registros).
    """

    # Abrimos el archivo en modo lectura
    with open(ruta, "r", encoding="utf-8") as archivo:
        lineas = archivo.readlines()

    # Primera línea → nombres de columnas
    encabezados = lineas[0].strip().split(",")

    # Lista que almacenará todos los registros
    datos = []

    # Definimos columnas numéricas del dataset clínico
    columnas_numericas = [
        "id", "age", "hypertension", "heart_disease",
        "avg_glucose_level", "bmi", "stroke"
    ]

    # Recorremos desde la segunda línea (datos reales)
    for linea in lineas[1:]:

        # Separar valores por coma
        valores = linea.strip().split(",")

        # Diccionario para representar un paciente
        fila = {}

        # Asociamos cada valor con su columna
        for i in range(len(encabezados)):

            valor = valores[i]
            columna = encabezados[i]

            # Manejo básico de valores faltantes
            # Si el valor está vacío o es "nan"
            # lo convertimos en None
            if valor == "" or valor.lower() == "nan":
                fila[columna] = None

            # Convertimos a float SOLO si es numérico
            elif columna in columnas_numericas:
                try:
                    fila[columna] = float(valor)
                except:
                    fila[columna] = None

            # Variables categóricas se mantienen como texto
            else:
                fila[columna] = valor

        # Agregamos el registro a la lista principal
        datos.append(fila)

    return encabezados, datos


# -----------------------------------------
# EJECUCIÓN DE LA FUNCIÓN
# -----------------------------------------
# Aquí comienza formalmente el pipeline.
# Cargamos el dataset clínico de stroke.

encabezados, datos = cargar_csv("healthcare-dataset-stroke-data.csv")

print("Registros cargados:", len(datos))
print("Columnas:", encabezados)
````

````python
# -----------------------------------------
# FUNCIÓN: contar_nulos
# -----------------------------------------
# Esta función pertenece a la etapa de
# VALIDACIÓN Y LIMPIEZA DE DATOS.
#
# Objetivo:
# - Detectar cuántos valores faltantes (None)
#   existen por cada columna.
#
# ¿Por qué es importante?
# - Permite evaluar calidad del dataset.
# - Es un paso previo obligatorio antes
#   de análisis estadístico o modelado.
# - Ayuda a decidir estrategias de imputación
#   o eliminación de registros.


def contar_nulos(datos):
    # Diccionario donde almacenaremos el conteo
    # Ejemplo esperado:
    # {"bmi": 10, "avg_glucose_level": 2}
    conteo = {}

    # Recorremos cada registro del dataset
    # Complejidad: O(n)
    for fila in datos:

        # Recorremos cada variable del registro
        # Complejidad interna: O(m)
        for clave, valor in fila.items():

            # Si el valor es None (faltante)
            if valor is None:

                # Si la columna aún no está en el diccionario
                # la inicializamos en 0
                if clave not in conteo:
                    conteo[clave] = 0

                # Incrementamos el contador
                conteo[clave] += 1

    return conteo


# -----------------------------------------
# EJECUCIÓN DE VALIDACIÓN
# -----------------------------------------
# Mostramos el resumen de valores faltantes
# detectados en el dataset clínico.

print("Valores faltantes:", contar_nulos(datos))
````

Limpieza de Datos

Se eliminan registros con valores no válidos.

````python
# -----------------------------------------
# FUNCIÓN: limpiar_datos
# -----------------------------------------
# Esta función corresponde a la etapa de
# LIMPIEZA FORMAL DEL DATASET.
#
# Objetivo:
# - Eliminar registros con datos inválidos
#   en variables críticas (avg_glucose_level y bmi).
#
# ¿Por qué estas variables?
# - Son variables centrales en el análisis.
# - Se utilizan en la función de decisión heurística.
# - Valores <= 0 no tienen sentido clínico.
#
# Estrategia aplicada:
# - Si hay None → descartar registro.
# - Si valor <= 0 → descartar registro.
#
# No se modifica el dataset original directamente.
# Se crea una nueva lista limpia.


def limpiar_datos(datos):
    """
    Elimina registros con valores inválidos básicos.

    Parámetro:
    datos (list): Dataset original.

    Retorna:
    list: Dataset filtrado.
    """

    # Nueva lista que almacenará solo registros válidos
    datos_limpios = []

    # Recorremos todos los registros
    # Complejidad: O(n)
    for fila in datos:

        #Eliminación por valores faltantes
        if fila["avg_glucose_level"] is None or fila["bmi"] is None:
            continue

        # Eliminación por valores físicamente inválidos
        if fila["avg_glucose_level"] <= 0 or fila["bmi"] <= 0:
            continue

        # Si pasa todas las validaciones, se conserva
        datos_limpios.append(fila)

    return datos_limpios


# -----------------------------------------
# EJECUCIÓN DE LA LIMPIEZA
# -----------------------------------------

datos = limpiar_datos(datos)

print("Registros tras limpieza:", len(datos))
````

Cálculo de Promedios

````python
# -----------------------------------------
# FUNCIÓN: calcular_promedios
# -----------------------------------------
# Esta función pertenece a la etapa de
# ANÁLISIS DESCRIPTIVO DEL DATASET.
#
# Objetivo:
# - Calcular el promedio de múltiples variables
#   dentro de un grupo específico (por ejemplo:
#   pacientes con o sin stroke).
#
# Características importantes:
# - Ignora valores None (faltantes).
# - Evita división por cero.
# - Devuelve resultados estructurados en diccionario.
#
# Esto permite comparar estadísticamente grupos
# antes de aplicar modelos más complejos.


def calcular_promedios(grupo, variables):
    # Diccionario donde almacenaremos resultados
    resultados = {}

    # Recorremos cada variable solicitada
    # Complejidad externa: O(k)
    for var in variables:

        suma = 0
        contador = 0

        # Recorremos todos los registros del grupo
        # Complejidad interna: O(n)
        for fila in grupo:
            valor = fila[var]

            # Ignoramos valores faltantes y aseguramos tipo numérico
            if valor is not None and isinstance(valor, (int, float)):
                suma += valor
                contador += 1

        # Evitamos división por cero
        resultados[var] = suma / contador if contador > 0 else 0

    return resultados


# -----------------------------------------
# EJECUCIÓN DEL ANÁLISIS DESCRIPTIVO
# -----------------------------------------

variables = ["avg_glucose_level", "bmi", "age"]

# Calculamos promedios por grupo
prom_stroke = calcular_promedios(grupo_stroke, variables)
prom_no_stroke = calcular_promedios(grupo_no_stroke, variables)

print("Promedios Stroke:", prom_stroke)
print("Promedios No Stroke:", prom_no_stroke)
````

````python
# Diferencias entre grupos
diff = {var: prom_stroke[var] - prom_no_stroke[var] for var in variables}

print("Diferencias (Stroke - No Stroke):", diff)
````

Detección de Outliers Simples

````python
# -----------------------------------------
# FUNCIÓN: detectar_outliers
# -----------------------------------------
# Esta función pertenece a la etapa de
# VALIDACIÓN AVANZADA DEL DATASET.
#
# Objetivo:
# - Detectar registros con valores extremos
#   según reglas simples basadas en umbrales.
#
# Regla aplicada:
# - avg_glucose_level > 250
# - bmi > 60
#
# ⚠ Importante:
# - No elimina los registros.
# - Solo los identifica.
# - Permite análisis posterior.
#
# Este enfoque es determinístico y simple.
# No utiliza métodos estadísticos como:
# - IQR
# - Z-score
# - Desviación estándar
#
# Complejidad: O(n)


def detectar_outliers(datos):
    # Lista donde almacenaremos registros sospechosos
    outliers = []

    # Recorremos todos los registros
    for fila in datos:

        # Evaluamos condiciones extremas (validando None)
        if (
            (fila["avg_glucose_level"] is not None and fila["avg_glucose_level"] > 250)
            or
            (fila["bmi"] is not None and fila["bmi"] > 60)
        ):
            outliers.append(fila)

    return outliers


# -----------------------------------------
# EJECUCIÓN DE DETECCIÓN
# -----------------------------------------

print("Outliers detectados:", len(detectar_outliers(datos)))
````

Transformación (Escalamiento)

````python
# -----------------------------------------
# FUNCIÓN: escalar_glucose
# -----------------------------------------
# Esta función pertenece a la etapa de
# TRANSFORMACIÓN Y PREPROCESAMIENTO.
#
# Objetivo:
# - Escalar la variable "avg_glucose_level"
#   dividiéndola por 200.
#
# ¿Por qué 200?
# - Es un valor de referencia aproximado
#   para llevar la variable a un rango cercano a [0,1].
# - Permite reducir magnitudes grandes.
# - Facilita comparación entre variables.
#
# Importante:
# - No modifica el dataset original.
# - Crea una nueva estructura.
# - Aplica copia superficial del diccionario.
#
# Complejidad: O(n)


def escalar_glucose(datos):
    # Nueva lista donde almacenaremos los registros transformados
    datos_nuevos = []

    # Recorremos todos los registros
    for fila in datos:

        # Creamos copia del registro
        # para no modificar el original
        nueva = fila.copy()

        # Aplicamos transformación simple (validando None)
        if fila["avg_glucose_level"] is not None:
            nueva["avg_glucose_level"] = fila["avg_glucose_level"] / 200
        else:
            nueva["avg_glucose_level"] = None

        # Guardamos el registro transformado
        datos_nuevos.append(nueva)

    return datos_nuevos


# -----------------------------------------
# EJECUCIÓN DE TRANSFORMACIÓN
# -----------------------------------------

datos_escalados = escalar_glucose(datos)

# Validación estructural:
# Verificamos que no se hayan perdido registros
assert len(datos) == len(datos_escalados)

print("Escalamiento aplicado correctamente.")
````

Lectura con Pandas

````python
# -----------------------------------------
# LECTURA CON PANDAS
# -----------------------------------------
# Aquí comenzamos el ANEXO VECTORIZADO.
#
# Objetivo:
# - Mostrar cómo librerías especializadas
# simplifican el mismo pipeline.
#
# Diferencia clave:
# - pandas implementa estructuras tipo DataFrame
# - Las operaciones están optimizadas en C
# - Se reduce código explícito
# - Se mejora legibilidad

import pandas as pd
import numpy as np

# Lectura directa del CSV (sin loops manuales)
df = pd.read_csv("healthcare-dataset-stroke-data.csv")

# shape devuelve (filas, columnas)
print("Dimensiones:", df.shape)
````

Limpieza con Pandas

````python
# -----------------------------------------
# LIMPIEZA CON PANDAS
# -----------------------------------------
# Equivalente a nuestra función limpiar_datos(),
# pero ahora usando operaciones vectorizadas.
#
# dropna elimina registros con valores nulos
# en columnas específicas.
#
# Luego aplicamos filtro booleano en una sola línea.

df_limpio = df.dropna(subset=["avg_glucose_level", "bmi"])

df_limpio = df_limpio[
    (df_limpio["avg_glucose_level"] > 0) &
    (df_limpio["bmi"] > 0)
]

print("Registros tras limpieza:", len(df_limpio))
````

Promedios con groupby

````python
# -----------------------------------------
# PROMEDIOS CON GROUPBY
# -----------------------------------------
# Equivalente a calcular_promedios()
# pero usando agrupamiento vectorizado.
#
# groupby("stroke") separa automáticamente
# los grupos 0 y 1.
#
# mean() calcula promedio sin loops explícitos.

variables = ["avg_glucose_level", "bmi", "age"]

promedios = df_limpio.groupby("stroke")[variables].mean()

print(promedios)
````

Detección de Outliers con IQR

````python
# -----------------------------------------
# DETECCIÓN DE OUTLIERS CON IQR
# -----------------------------------------
# Ahora utilizamos un método estadístico formal.
#
# IQR (Interquartile Range):
# IQR = Q3 - Q1
#
# Regla:
# Outlier si:
# x < Q1 - 1.5*IQR
# x > Q3 + 1.5*IQR
#
# Esto es más robusto que usar umbrales fijos.

Q1 = df_limpio["avg_glucose_level"].quantile(0.25)
Q3 = df_limpio["avg_glucose_level"].quantile(0.75)

IQR = Q3 - Q1

lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

outliers = df_limpio[
    (df_limpio["avg_glucose_level"] < lim_inf) |
    (df_limpio["avg_glucose_level"] > lim_sup)
]

print("Outliers detectados:", len(outliers))
````
