```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv("Titanic.csv")
df.head()
```

```python
df.columns

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Contamos la cantidad de sobrevivientes y no sobrevivientes por clase (Pclass)
class_survival = df.groupby(['Pclass', 'Survived']).size().unstack(fill_value=0)

# Crear gráfico de barras normales
class_survival.plot(kind='bar', stacked=False, color=['salmon', 'lightgreen'], figsize=(8, 6))

# Personalizar el gráfico
plt.title('Supervivencia por Clase')
plt.xlabel('Clase')
plt.ylabel('Número de Supervivientes y No Sobrevivientes')
plt.xticks(rotation=0)
plt.legend(['No sobrevivió', 'Sobrevivió'])
plt.show()

```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%1.1f%%', labels=['No sobrevivió', 'Sobrevivió'])
plt.title('Proporción de Supervivientes')
plt.ylabel('')  # Eliminar etiqueta en el eje Y
plt.show()
```

```python
# Agrupar los datos por 'Sex' y 'Survived' para contar los sobrevivientes por género
survivors_by_gender = df[df['Survived'] == 1].groupby('Sex').size().reset_index(name='Survivors')

# Crear gráfico de barras
sns.barplot(x='Sex', y='Survivors', data=survivors_by_gender)
plt.title('Número de Supervivientes por Género')
plt.xlabel('Género')
plt.ylabel('Número de Supervivientes')
plt.show()
```
